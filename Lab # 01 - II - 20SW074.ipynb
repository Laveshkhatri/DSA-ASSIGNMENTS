```python
string = "Hello Google"

print("capitalize():", string.capitalize())

print("casefold():", string.casefold())

# center(width[, fillchar])
print("center():", string.center(20, '-'))

# count(sub[, start[, end]])
print("count('o'):", string.count('o'))

# endswith(suffix[, start[, end]])
print("endswith('World'):", string.endswith('World'))

# find(sub[, start[, end]])
print("find('o'):", string.find('o'))

# index(sub[, start[, end]])
print("index('o'):", string.index('o'))

# isalnum()
print("isalnum():", string.isalnum())

# isalpha()
print("isalpha():", string.isalpha())

# isdigit()
print("isdigit():", string.isdigit())

# islower()
print("islower():", string.islower())

# isnumeric()
print("isnumeric():", string.isnumeric())

# isspace()
print("isspace():", string.isspace())

# istitle()
print("istitle():", string.istitle())

# isupper()
print("isupper():", string.isupper())

# join(iterable)
words = ['Hello', 'World']
print("join():", '-'.join(words))

# lower()
print("lower():", string.lower())

# lstrip([chars])
print("lstrip():", string.lstrip('Hello'))

# replace(old, new[, count])
print("replace('Hello', 'Hi'):", string.replace('Hello', 'Hi'))

# split([sep[, maxsplit]])
print("split():", string.split())

# startswith(prefix[, start[, end]])
print("startswith('Hello'):", string.startswith('Hello'))

# strip([chars])
print("strip():", string.strip('!'))

# swapcase()
print("swapcase():", string.swapcase())

# title()
print("title():", string.title())

# upper()
print("upper():", string.upper())
```

```python
students = {
    'John': {
        'age': 20,
        'major': 'Computer Science',
        'GPA': 3.7
    },
    'Emily': {
        'age': 19,
        'major': 'Mathematics',
        'GPA': 3.9
    },
    'Sarah': {
        'age': 21,
        'major': 'Physics',
        'GPA': 3.5
    }
}

# Accessing student data
print(students['John']['age'])  # Output: 20
print(students['Emily']['major'])  # Output: Mathematics
print(students['Sarah']['GPA'])  # Output: 3.5

# Modifying student data
students['John']['GPA'] = 3.8
students['Emily']['major'] = 'Computer Science'

# Adding a new student
students['Michael'] = {
    'age': 22,
    'major': 'Chemistry',
    'GPA': 3.6
}

# Deleting a student
del students['Sarah']

# Printing all student data
for student, data in students.items():
    print(student)
    for key, value in data.items():
        print(f"{key}: {value}")
    print()
```
