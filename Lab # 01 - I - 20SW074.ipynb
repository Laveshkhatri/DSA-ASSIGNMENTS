```python
#15 Generic Operations in Containers

lst = [11,14,16,18,19]
lost = []
for idx in range(len(lst)):
  val = lst[idx]
  if val > 15:
    lost.append(val)
    lst[idx] = 15
print("modif: ",lst, "-lost: ",lost)


for idx,val in enumerate(lst):
  print(idx,"----",val)
```
